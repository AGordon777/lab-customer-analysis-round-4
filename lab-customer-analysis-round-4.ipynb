```python
# scale the feature columns using MinMaxScaler() transform
from sklearn.preprocessing import MinMaxScaler  # i'm going to use this one
from sklearn.preprocessing import StandardScaler
```

```python
numerical.describe(include='all').T
```

```python
#looking at the data description above, customer lifetime value looks to have the biggest tail
# max is around 10x bigger than the 75% value, so there must be a large tail
# i am aware that we may not want to exclude so valuable a group of customers!!!
```

```python
sns.boxplot(y='Customer Lifetime Value', data=data)
plt.show()
```

```python
#so let's take out the outliers
```

```python
Customer_Lifetime_Value_iqr = np.percentile(numerical['Customer Lifetime Value'],75) - np.percentile(numerical['Customer Lifetime Value'],25)
Customer_Lifetime_Value_upper_limit = np.percentile(numerical['Customer Lifetime Value'],75) + 1.5*Customer_Lifetime_Value_iqr
Customer_Lifetime_Value_lower_limit = np.percentile(numerical['Customer Lifetime Value'],25) - 1.5*Customer_Lifetime_Value_iqr

numerical_filtered = numerical[(numerical['Customer Lifetime Value']>Customer_Lifetime_Value_lower_limit) & (numerical['Customer Lifetime Value']<Customer_Lifetime_Value_upper_limit)].copy()
numerical_filtered
```

```python
# i've taken out (9134 - 8317) = 817 rows (9% of data)
# i could continue with other columns, but don't want to loe too much data yet
```

```python
# assign the target, and the remaining columns are the features (we have not dropped any)
y = numerical_filtered['Total Claim Amount'] # this is the target
X = numerical_filtered.drop(['Total Claim Amount'], axis=1) # this is the array of features
```

```python
#as a first pass scaling exercise, i'm just seeing what the data looks like
#i'm not removing outliers at this stage

transformer = MinMaxScaler().fit(X)
x_standardized = transformer.transform(X)
X_scaled = pd.DataFrame(x_standardized, columns=X.columns)
X_scaled
```
